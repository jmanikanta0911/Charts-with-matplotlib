```python
# Import pandas and read the csv file
import pandas as pd
data = pd.read_csv("/content/cri t20.csv")
```

```python
# Display the first ten rows
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
x = data.Player
y = data.Runs
plt.fill_between(x,y)
plt.title("Area Chart")
plt.ylabel("Runs")
plt.xticks(rotation='vertical')
plt.xlabel("Player")
plt.show()
```

```python
import matplotlib.pyplot as plt
items = [data.Runs,data.Ave,data.SR]
plt.boxplot(items)
plt.title("Box plot")
plt.show()
```
